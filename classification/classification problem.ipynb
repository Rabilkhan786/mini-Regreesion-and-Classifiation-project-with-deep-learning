### Loading Data set

```python
import torchvision
import torch
from torchvision import transforms
from torch.utils.data import DataLoader
import torch.nn as nn


Image_path = './'
transform = transforms.Compose([
    transforms.ToTensor()
])

mnist_train_dataset = torchvision.datasets.MNIST(
    root='Image_path', train = True,
    transform = transform, 
    download = True
)

mnist_test_dataset = torchvision.datasets.MNIST(
    root='Image_path', train= False,
    transform=transform, download=False
)


batch_size = 64
torch.manual_seed(1)
train_dl = DataLoader(mnist_test_dataset,batch_size=batch_size, shuffle=True)
```

```python
x_batch, y_batch = next(iter(train_dl))

print(x_batch.shape)
print(y_batch.shape)
```

## Moddel building for this classification problem 

```python
hidden_units = [32,16]
image_size = mnist_train_dataset[0][0].shape
input_size = image_size[0] * image_size[1] * image_size[2]

all_layers = [nn.Flatten()]

for hidden_unit in hidden_units:
   layer = nn.Linear(input_size, hidden_unit)
   all_layers.append(layer)
   all_layers.append(nn.ReLU())
   input_size = hidden_unit
all_layers.append(nn.Linear(hidden_units[-1],10))
model = nn.Sequential(*all_layers)
```

```python
model
```

### Training the model

```python
loss_fn = nn.CrossEntropyLoss()
optimizer =torch.optim.Adam(model.parameters(),lr=0.001) # using adam optimizer for best result

torch.manual_seed(1) 
num_epochs = 20

for epoch in range(num_epochs):
    accuracy_hist_train = 0
    for x_batch, y_batch in train_dl:
        # step 1
        pred = model(x_batch)
        # step 2
        loss = loss_fn(pred,y_batch)
        # step 3
        loss.backward()
        # step 4
        optimizer.step()
        # step 5
        optimizer.zero_grad()
        is_correct = (
            torch.argmax(pred,dim=1) == y_batch
        ).float()
        # storing accuracy for every epoch 
        accuracy_hist_train += is_correct.sum()
    accuracy_hist_train /= len(train_dl.dataset)
    print(f'Epoch {epoch} Accuracy '
          f'{accuracy_hist_train:.4f} ')
    
    
```

### Evaluate the model

```python
pred = model(mnist_test_dataset.data/255.) # we suing 8-bit pixel image . so it is normalized by 255

is_correct = (
    torch.argmax(pred,dim=1) == mnist_test_dataset.targets
).float()

print(f'Test accuracy: {is_correct.mean():.4f}')
```
